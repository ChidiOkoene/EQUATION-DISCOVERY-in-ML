```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import RFECV
from sklearn.model_selection import train_test_split

def feature_selection_with_rfecv(X, y, model):
    """
    Perform Recursive Feature Elimination with Cross-Validation (RFECV) to select important features.
    
    Parameters:
    - X: pd.DataFrame, Feature matrix
    - y: pd.Series, Continuous target variable
    - model: sklearn model, The model to use for RFECV (e.g., RandomForestRegressor)
    
    Returns:
    - selected_features: list, Names of the selected features
    - feature_ranking: list, Ranking of all features (1 means selected)
    - optimal_number_of_features: int, Optimal number of features selected by RFECV
    """
    
    # Initialize RFECV with cross-validation
    rfecv = RFECV(estimator=model, step=1, cv=5, scoring='neg_mean_squared_error', min_features_to_select=20)  
    
    # Fit RFECV
    rfecv.fit(X, y)
    
    # Get the selected features
    selected_features = list(X.columns[rfecv.support_])
    
    # Get the ranking of features
    feature_ranking = list(rfecv.ranking_)
    
    # Get the optimal number of features
    optimal_number_of_features = rfecv.n_features_
    
    return selected_features, feature_ranking, optimal_number_of_features

```

```python
selected_features, feature_ranking, optimal_number_of_features = feature_selection_with_rfecv(X, y, RandomForestRegressor())
```
